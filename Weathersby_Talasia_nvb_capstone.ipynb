```python
# ==========================================
# Warehouse Inventory Management System
# NVB Python Capstone - Summer 2026
# ==========================================

# Dictionary to store inventory
inventory = {
    "Bolts": 150,
    "Nuts": 45,
    "Washers": 60,
    "Screws": 120
}


def display_inventory():
    """
    Display all inventory items and their quantities.
    """
    print("\n========== CURRENT INVENTORY ==========")

    for item, quantity in inventory.items():
        print(f"{item}: {quantity}")

    print("=======================================\n")


def add_stock():
    """
    Add a new inventory item or restock an existing item.
    """
    item = input("Enter item name: ").title()

    try:
        quantity = int(input("Enter quantity to add: "))

        if quantity <= 0:
            print("Quantity must be greater than zero.\n")
            return

        if item in inventory:
            inventory[item] += quantity
            print(f"{quantity} added to {item}.")
        else:
            inventory[item] = quantity
            print(f"{item} added to inventory.")

    except ValueError:
        print("Please enter a valid number.")


def process_order():
    """
    Process a customer order by reducing inventory.
    """
    item = input("Enter item to order: ").title()

    if item not in inventory:
        print("Item not found.\n")
        return

    try:
        quantity = int(input("Enter quantity ordered: "))

        if quantity <= 0:
            print("Quantity must be greater than zero.")
            return

        if quantity > inventory[item]:
            print("Not enough stock available.")
            return

        inventory[item] -= quantity

        print(f"Order processed for {quantity} {item}.")

        if inventory[item] < 10:
            print(f"⚠ LOW STOCK ALERT: {item} only has {inventory[item]} left!")

    except ValueError:
        print("Please enter a valid number.")


def low_stock_report():
    """
    Display all items with fewer than 10 units.
    """
    print("\n========== LOW STOCK REPORT ==========")

    found = False

    for item, quantity in inventory.items():
        if quantity < 10:
            print(f"{item}: {quantity}")
            found = True

    if not found:
        print("No low-stock items.")

    print("======================================\n")


def menu():
    """
    Display the main menu.
    """
    print("\n===================================")
    print(" Warehouse Inventory Management")
    print("===================================")
    print("1. View Inventory")
    print("2. Add Stock")
    print("3. Process Order")
    print("4. Low Stock Report")
    print("5. Exit")


# Main Program Loop
while True:

    menu()

    choice = input("Choose an option (1-5): ")

    if choice == "1":
        display_inventory()

    elif choice == "2":
        add_stock()

    elif choice == "3":
        process_order()

    elif choice == "4":
        low_stock_report()

    elif choice == "5":
        print("\nThank you for using the Warehouse Inventory Management System!")
        break

    else:
        print("Invalid option. Please choose 1-5.")
```
